```python
def cleaner(text):
    li=[]
    for var in text:
        if (var !='(c)') and (var !='(vc)') and (var !='(wk)') and (var !='(vc, wk)') and (var !='(c, wk)'):
            li.append(var)
        else:
            pass
    return li
```

```python
def player_cleaner(df):
    for var in country_name:
        df[var]=df[var].fillna('0').str.split().apply(cleaner).apply(lambda x:' '.join(x))
    return df
```

## Extracting total ball thrown

```python
bowler['co']=bowler['Overs'].str.extract('(\d+)')
bowler['uo']=bowler['Overs'].str.extract('(\.\d+)')
bowler['co'].fillna('0',inplace=True)
bowler['uo'].fillna('0',inplace=True)
bowler['co']=bowler['co'].astype('int')
bowler['uo']=bowler['uo'].astype('float')
bowler['uo']=bowler['uo']*10
bowler['uo']=bowler['uo'].astype('int')
bowler['ball_thrown']=bowler['co']*6+bowler['uo']
bowler.drop(['co','uo'],axis=1,inplace=True)
```
